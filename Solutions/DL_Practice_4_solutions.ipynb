## Width vs Depth

Our goal here is to compare the performances of basic networks.
We will create both very wide and very deep networks, and see which ones are better.

We will try to fit a sequence of functions with increasing complexity, both with a wide and with a deep network.
The first part concentrates on the minimal amount of neurons needed to fit the function with an optimal network (setting the weights and biases manually).
Then, the second part studies the training of the same networks, to fit the same functions. 

### Theory

Taking the function of interest: $f: \mathbb{R} \to \mathbb{R}$ to be linear by segment, with 4 segments:
- $f(x) = 0$ on $\left] -\infty, 0 \right]$
- $f(x) = 2x$ on $\left] 0, \frac{1}{2} \right]$
- $f(x) = 2-2x$ on $\left] \frac{1}{2}, 1 \right]$
- $f(x) = 0$ on $\left] 1, \infty \right]$

Define $f : x\: \rightarrow f(x)$ as a python function, using numpy. It's the same as $f(x) = 1 - min(2|x - 0.5|, 1)$. You can prove it easily.

Let also:
- $g(x, 2) = f \circ f(x)$
- $g(x, 3) = f \circ f \circ f(x)$
- $g(x, 4) = f \circ f \circ f \circ f(x)$
- etc...

Define $g : x\: \rightarrow g(x, l)$ as a python function for all $l \in \mathbb{N}^*$.

Plot $f$ and $g$ on $\left[ -0.2, 1.2 \right]$

```python
def f(x):
    return 1 - np.minimum(2 * np.abs(x - 0.5), 1)

x = np.linspace(-0.2, 1.2, 10000)
y = f(x)
plt.plot(x, y, label="f(x)")
y = f(f(x))
plt.plot(x, y, label="f(f(x))")

def g(x, lvl=5):
    for i in range(lvl):
        x = f(x)
    return x

y = g(x)
plt.plot(x, y, label="g(x)")
plt.legend()
plt.show()
```

```python
class Net(nn.Module):
    def __init__(self, n_layers=1, n_hidden=3):
        super(Net, self).__init__()
        self.fc_in = nn.Linear(1, n_hidden)
        self.fcs = []
        for i in range(n_layers-1):
            self.fcs.append(nn.Linear(n_hidden, n_hidden))
        self.fc_out = nn.Linear(n_hidden, 1)

    def forward(self, x):
        x = nn.ReLU()(self.fc_in(x))
        for i in range(len(self.fcs)):
            x = nn.ReLU()(self.fcs[i](x))
        x = self.fc_out(x)
        return x
```

Formal proof of why it works (for $f(f(x))$):

Recall that we represent $f(x)$ as:$$f(x) = V \cdot \sigma(W x + b)$$
Where:
- $W = \begin{bmatrix} 1 \\ 1 \\ 1 \end{bmatrix}$
- $b = \begin{bmatrix} 0 \\ -0.5 \\ -1 \end{bmatrix}$
- $V = \begin{bmatrix} 2 & -4 & 2 \end{bmatrix}$

Now, let's look at what happens when we want to compute $z = f(f(x))$:

We take the output of the first $f$ and use it as the input for the second $f$.$$z = \underbrace{V \cdot \sigma(W \cdot \underbrace{(V \cdot \sigma(W x + b))}_{f(x)} + b)}_{\text{Second } f}$$

We have a $W \cdot V$, which will give us the matrix $M$ defined above:

$$M = W \cdot V = \begin{bmatrix} 1 \\ 1 \\ 1 \end{bmatrix} \begin{bmatrix} 2 & -4 & 2 \end{bmatrix} = \begin{bmatrix} 2 & -4 & 2 \\ 2 & -4 & 2 \\ 2 & -4 & 2 \end{bmatrix}$$

So all in all we have:
$$z = V \cdot \sigma(M \cdot \sigma(W x + b) + b)$$

### Training

We will try to fit `g` with `level = 4` both with deep and wide networks; This time, by training the network, instead of manually setting the weights.

First, train a wide network, of course, it will need more neurons than the mathematically optimal solutions.
Try with 5 times more neurons than in the optimal solution, and about $50000$ epochs.

```python
# create a wide network for lvl=5 and train it
level = 4
n = 2**level + 1
print(f'Neurons used (theory): {n}')
wide = Net(0, n*5)
print(f'Neurons used (practice): {(n*5)}')

x = np.linspace(-0.2, 1.2, 1000)
y = g(x, level)

# train the network
loss_fn = torch.nn.MSELoss()
optimizer = torch.optim.Adam(wide.parameters(), lr=0.02)
for i in range(50000):
    optimizer.zero_grad()
    pred = wide(torch.tensor(x).float().view(-1, 1))
    loss = loss_fn(pred, torch.tensor(y).float().view(-1, 1))
    loss.backward()
    optimizer.step()
    if i % 1000 == 0:
        print(f'Loss: {loss.item()}')

# plot the result
y = wide(torch.tensor(x).float().view(-1, 1)).detach().numpy()
plt.plot(x, y, label="wide(x)")
```

Now, train a deep network, again, give a little slack on the number of neurons compared to the optimal solution.
Try with 5 times more layers than necessary, and 15 neurons per layer instead of 3, and about $50000$ epochs.

```python
# cretae a deep network and train it
level = 4
print(f'Neurons used (theory): {3*level}')
deep = Net(level*5, 15)
print(f'Neurons used (practice): {15*(level*5)}')

x = np.linspace(-0.2, 1.2, 1000)
y = g(x, level)

# train the network
loss_fn = torch.nn.MSELoss()
optimizer = torch.optim.Adam(deep.parameters(), lr=0.001)
for i in range(50000):
    optimizer.zero_grad()
    pred = deep(torch.tensor(x).float().view(-1, 1))
    loss = loss_fn(pred, torch.tensor(y).float().view(-1, 1))
    loss.backward()
    optimizer.step()
    if i % 1000 == 0:
        print(f'Loss: {loss.item()}')

# plot the result
y = deep(torch.tensor(x).float().view(-1, 1)).detach().numpy()
plt.plot(x, y, label="deep(x)")
```

**Conclusion:**

The number of neurons is a good measure of the size (in terms of memory) of your network.

For the same amount of neurons, deep networks, can catch more complexity, but are harder to train.
On the contrary, wide networks catch less complexity, but are easier to train.

*Your goal as an AI engineer is to find the best architectures, so that your networks are both trainable and able to catch the complexity of the observed phenomenon.*
